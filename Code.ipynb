```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
act = pd.read_csv('dailyActivity_merged.csv')
act
```

```python
sleep = pd.read_csv('sleepDay_merged.csv')
sleep
```

```python
weight = pd.read_csv('weightLogInfo_merged.csv')
weight
```

```python
weight = weight.drop(columns = ['WeightKg' , 'Fat', 'LogId', 'IsManualReport', 'Date'])
weight = weight.groupby(by = ['Id']).mean()
weight = weight.reset_index()
weight
```

```python
activity = act.merge(weight, on = ['Id'], how = 'left')
activity
```

```python
step_id = activity.groupby(by = 'Id').mean()[['TotalSteps']]
step_id
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(data=activity, x = 'TotalSteps')
plt.title("Boxplot of total steps")
plt.show()
```

```python
active_status = []
for s in step_id['TotalSteps']:
    if s < 7000:
        active_status.append('not active')
    elif s >= 7000 and s <= 10000:
        active_status.append('moderate active')
    elif s> 10000:
        active_status.append('very active')
step_id['ActiveStatus'] = active_status
step_id
```

```python
percent_active = step_id.groupby(by = ['ActiveStatus']).count()
percent_active = percent_active.rename(columns = {'TotalSteps': 'Count'})
percent_active['Total'] = np.sum(percent_active['Count'])
percent_active['% of Status'] = (percent_active['Count'] / percent_active['Total']) * 100
percent_active
```

```python
percent_active = percent_active.reset_index()
plt.pie(percent_active['% of Status'], labels = percent_active['ActiveStatus'])
plt.title('User Type by Active Status')
plt.show() 
```

Most of the users are on lightly active status, so only 20% of users are following CDC Guideline. 

```python
activity['ActivityDate'] = pd.to_datetime(activity['ActivityDate'])
activity['Day'] = activity['ActivityDate'].dt.dayofweek
#activity['Day'] = activity['Day'].replace({0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'})
activity
```

```python
day_act = activity.groupby(by = ['Day']).mean()
day_act = day_act[['TotalSteps', 'TotalDistance', 'Calories']]
day_act = day_act.reset_index()
day_act['Day'] = day_act['Day'].replace({0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'})
day_act
```

```python
plt.bar(day_act['Day'], day_act['TotalSteps'])
plt.xticks(rotation = 30)
plt.xlabel("Day")
plt.ylabel("Average Steps")
plt.title("Average Steps by Day")
plt.show()
```

```python
plt.bar(day_act['Day'], day_act['TotalDistance'])
plt.xticks(rotation = 30)
plt.xlabel("Day")
plt.ylabel("Average Distance")
plt.title("Average Distance by Day")
plt.show()
```

```python
plt.bar(day_act['Day'], day_act['Calories'])
plt.xticks(rotation = 30)
plt.xlabel("Day")
plt.ylabel("Average Calories")
plt.title("Average Calories by Day")
plt.show()
```

Users tend to walk farther ditance on Tuesday and Saturday

```python
sleep['TotalMinutesNotSleep'] = sleep['TotalTimeInBed'] - sleep['TotalMinutesAsleep']
sleep['% of NotSleep'] = (sleep['TotalMinutesNotSleep'] / sleep['TotalTimeInBed']) * 100
sleep['Hours Sleep'] = sleep['TotalMinutesAsleep'] / 60
sleep
```

```python
sleep_id = sleep.groupby(by = 'Id').mean()[['Hours Sleep']]
sleep_id
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(data=sleep, x = 'Hours Sleep')
plt.title("Boxplot of # of Hours Asleep")
plt.show()
```

```python
sleep_status = []
for t in sleep_id['Hours Sleep']: 
    if t >= 7.0:
        sleep_status.append('Enough Sleep')
    else:
        sleep_status.append('Not Enough Sleep')
sleep_id['Sleep Status'] = sleep_status
sleep_id
```

```python
sleep_id = sleep_id.reset_index()
percent_sleep = sleep_id.groupby(by = ['Sleep Status']).count()
percent_sleep = percent_sleep.rename(columns = {'Hours Sleep': 'Count'})
percent_sleep['Total'] = np.sum(percent_sleep['Count'])
percent_sleep['% of Status'] = (percent_sleep['Count'] / percent_sleep['Total']) * 100
percent_sleep
```

```python
percent_sleep = percent_sleep.reset_index()
plt.pie(percent_sleep['% of Status'], labels = percent_sleep['Sleep Status'])
plt.title('User Type by Sleep Hours')
plt.show() 
```

According to CDC guideline, people should sleep 7 hours or more. Based on our data, only 50% of users sleep 7 ours or more.

```python
sleep['SleepDay'] = pd.to_datetime(sleep['SleepDay'])
sleep['Day'] = sleep['SleepDay'].dt.dayofweek
sleep
```

```python
sleep_day = sleep.groupby(by = 'Day').mean()[['Hours Sleep']]
sleep_day = sleep_day.reset_index()
sleep_day['Day'] = sleep_day['Day'].replace({0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'})
sleep_day
```

```python
plt.bar(sleep_day['Day'], sleep_day['Hours Sleep'])
plt.xticks(rotation = 30)
plt.xlabel('Day')
plt.ylabel('# of Hours Sleep')
plt.title('Average # of Hours Sleep by Day')
```

People tend to sleep longer on Wednesday and Sunday, where users spend burn more calories the day before. 

```python
no_sleep = sleep.groupby(by = 'Day').mean()[['TotalMinutesNotSleep']]
no_sleep = no_sleep.reset_index()
no_sleep['Day'] = no_sleep['Day'].replace({0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'})
no_sleep
```

```python
plt.bar(no_sleep['Day'], no_sleep['TotalMinutesNotSleep'])
plt.xticks(rotation = 30)
plt.xlabel('Day')
plt.ylabel('# of Minutes Needed to be Spent to get Asleep')
plt.title('Average # of Minutes Needed to be Spent to get Asleep')
```

```python
in_bed = sleep.groupby(by = 'Day').mean()[['TotalTimeInBed']]
in_bed = in_bed.reset_index()
in_bed['Day'] = in_bed['Day'].replace({0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'})
in_bed['Hours In Bed'] = in_bed['TotalTimeInBed'] / 60
in_bed
```

```python
plt.bar(in_bed['Day'], in_bed['Hours In Bed'])
plt.xticks(rotation = 30)
plt.xlabel('Day')
plt.ylabel('# of Hours In Bed')
plt.title('Average # of Hours In Bed')
```
